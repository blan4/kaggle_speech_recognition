```python
import sound_reader as sr
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from IPython.display import Audio
```

```python
L = 16000
ms_to_s = 1000.0
frame_size = int((20 / ms_to_s) * L)  # 320
stride_size = int((6.25 / ms_to_s) * L)  # 100
strides = int(L / stride_size)  # 160
frame_size, stride_size, strides
```

```python
wav = sr.read_wav_file('../data/train/audio/no/0132a06d_nohash_1.wav')
print(wav.shape)
plt.plot(wav)
Audio(wav-0.5, rate=L)
```

```python
w = np.pad(wav, frame_size//2, mode='constant', constant_values=0.5)

ww = np.vstack([w[i * stride_size:i*stride_size+frame_size].reshape(1, frame_size) for i in range(strides)])
print(ww.shape)
plt.imshow(ww, cmap='gray')
```
