```python
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../submissions/2017-11-18_20-29_confusion_matrix.csv', index_col='index')
df.head(12)
```

```python
cm = df.values
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(12,12))

df2 = pd.DataFrame(cm, columns=df.columns, index=df.index)

sns.heatmap(df2, cmap=plt.cm.Blues, cbar=False, annot=df.values, square=True, fmt='.4g')
```
